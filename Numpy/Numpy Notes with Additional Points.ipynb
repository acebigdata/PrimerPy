### Additional Array Indexing and Slicing

```python
arr_2d = np.zeros((10,10))
arr_2d
arr_shape = arr_2d.shape
arr_len = arr_shape[1]
arr_2d_copy = arr_2d.copy()
```

```python
for i in range(arr_len):
    arr_2d_copy[i] = i
arr_2d_copy
```

```python
for j in range(arr_len):
    arr_2d_copy[:,j] = j
arr_2d_copy
```

```python
arr_2d_3 = np.array(range(100)).reshape((10,10))
arr_2d_3
```

```python
for i,j in zip(range(arr_len),range(arr_len)):
    arr_2d_copy[i][j] = arr_2d_3[i][j]
arr_2d_copy
```

```python
for i,j in zip(range(arr_len),range(arr_len)):
    arr_2d_copy[i][j] = arr_2d_3[j][i]
arr_2d_copy
```

```python
for i in range(arr_len):
    for j in range(arr_len):
        arr_2d_copy[i][j] = arr_2d_3[i][j]

arr_2d_copy
```

```python
## List Comprehension
arr_A = np.array(range(1,5))
print(arr_A)
arr_B = arr_A * 100
print(arr_B)
arr_cond = np.array([True, True, False, False])
print(arr_cond)
```

```python
answer = [arr_A[arr_cond], arr_B[np.logical_not(arr_cond)]]
answer = np.array(answer)
answer = answer.reshape(4,)
answer
```

```python
## Alternatively
## Whole answer is an array
answer_2 = [A_val if cond_val else B_val for A_val, B_val, cond_val in zip(arr_A, arr_B, arr_cond)]
answer_2
answer_2 = np.array(answer_2) 
answer_2
```

```python
## Alternatively can use np.where
answer_3 = np.where(arr_cond, arr_A, arr_B)
answer_3
```
