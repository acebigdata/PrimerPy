## 3.1 Convert Data types for storage optimization 

```python
df = pd.read_csv("employees.csv")
df.head(3)
#conver to date time
df["Start Date"] = pd.to_datetime(df["Start Date"])
df["Last Login Time"] = pd.to_datetime(df["Last Login Time"])
#conver to boolean
df["Senior Management"] = df["Senior Management"].astype("bool")
#convert to category
df["Team"] = df["Team"].astype("category")
df["Gender"] = df["Gender"].astype("category")
```

## 3.3 Filter with More than One Condition (And "&", OR "|") 

```python
#conver to date time can be
#alternatively done via parse_dates
df_2 = pd.read_csv("employees.csv", parse_dates=["Start Date", "Last Login Time"])
#conver to boolean
df_2["Senior Management"] = df["Senior Management"].astype("bool")
#convert to category
df_2["Team"] = df["Team"].astype("category")
df_2["Gender"] = df["Gender"].astype("category")
df_2.head(3)
```

```python
# Create filters
gender_male = df_2["Gender"] == "Male"
team_marketing = df_2["Team"] == "Marketing"
```

```python
# use & not and
df_2[gender_male & team_marketing]
```

```python
senior_mgmt = df_2["Senior Management"]
high_salary = df_2["Salary"] >= 1000000
```

```python
df_2[senior_mgmt | high_salary]
```
