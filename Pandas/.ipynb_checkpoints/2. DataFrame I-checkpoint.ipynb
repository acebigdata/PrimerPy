## 2.9 Fill in Null Value with the .fillna() Method 

```python
nba_origin = pd.read_csv("nba.csv")
nba_origin.head(3)
```

```python
nba_origin.fillna(0)
# Doesn't make much sense on the string columns
```

```python
nba_origin["Salary"].mean()
```

```python
# Select single series and replace the series only
nba_origin["Salary"].fillna(nba_origin["Salary"].mean(), inplace=True)
```

```python
nba_origin
```

```python
nba_origin["College"].fillna("No College", inplace=True)
```

```python
nba_origin
```

```python
nba_origin = pd.read_csv("nba.csv").dropna(how="all")
nba_origin["Salary"].fillna(0, inplace=True)
nba_origin["College"].fillna("None", inplace=True)
nba_origin.head(3)
```

```python
nba_origin.dtypes
```

```python
nba_origin.info()
```

```python
nba_origin["Salary"] = nba_origin["Salary"].astype("int")
```

```python
nba_origin.head()
```

```python
nba_origin["Number"] = nba_origin["Number"].astype("int")
```

```python
nba_origin["Age"] = nba_origin["Age"].astype("int")
```

```python
nba_origin.head(3)
```

```python
nba_origin["Position"].nunique()
```

```python
#Category will decrease the memory use and increase the speed
nba_origin["Position"] = nba_origin["Position"].astype("category")
```

```python
nba_origin["Team"] = nba_origin["Team"].astype("category")
```

## 3.2 Rank Values with the .rank() Method 

```python
nba_origin = pd.read_csv("nba.csv")
nba_origin.head(3)
```

```python
nba_origin.dropna(how="all")
nba_origin["Salary"] = nba_origin["Salary"].fillna(0).astype("int")
nba_origin.head(3)
```

```python
nba_origin["Salary Rank"] = nba_origin["Salary"].rank(ascending=False).astype("int")
```

```python
nba_origin
```

```python
nba_origin.sort("Salary", ascending=False)
```

```python
nba_origin.sort("Salary Rank")
```
