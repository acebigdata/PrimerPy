```python
import pandas as pd
import numpy as np
```

```python
chicago = pd.read_csv("chicago.csv").dropna(how="all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.tail(3)
```

```python
chicago.info()
```

## 5.1 .lower(), .upper(), .title() and .len() 

```python
# Cannot use .title() on series directly
chicago["Name"].title()
```

```python
# Need to use .str first
chicago["Name"].str.title()
```

```python
chicago["Name"].str.lower().str.upper()
```

```python
chicago["Name"].apply(lambda name: str(name).title())
```

```python
chicago["Department"].str.len()
```

## 5.2 .str.replace() method 

```python
chicago["Department"] = chicago["Department"].str.replace("MGMNT", "Management").str.title()
```

```python
chicago["Department"].head()
```

```python
chicago["Employee Annual Salary"] = chicago["Employee Annual Salary"].str.replace("$","").astype(float)
```

```python
chicago["Employee Annual Salary"].mean()
```

```python
# Contains
mask = chicago["Position Title"].str.lower().str.contains("water")
```

```python
chicago[mask]
```

```python
#starts with water
mask2 = chicago["Position Title"].str.lower().str.startswith("water")
```

```python
chicago[mask2]
```

```python
#starts with water
mask3 = chicago["Position Title"].str.lower().str.endswith("ist")
```

```python
chicago[mask3]
```

## 5.3 Remove white spaces with .strip(), .strip() and .rstrip() Methods

```python
chicago = pd.read_csv("chicago.csv").dropna(how="all")
chicago["Department"] = chicago["Department"].astype("category")
chicago.tail(3)
```

```python
chicago["Name"] = chicago["Name"].str.strip()
```

```python
chicago
```

```python
chicago["Position Title"] = chicago["Position Title"].str.strip()
```
