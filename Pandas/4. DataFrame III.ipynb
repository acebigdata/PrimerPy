## 4.14 .apply() method on single columns 

```python
# sort index can retrieve much faster
bond_df = pd.read_csv("jamesbond.csv", index_col="Film")
bond_df.sort_index(inplace=True)
bond_df.head(3)
```

```python
def convert_to_string_and_add_millions(number):
    return str(number) + " Millsions"
```

```python
bond_df["Box Office"] = bond_df["Box Office"].apply(convert_to_string_and_add_millions)
```

```python
columns = ["Box Office", "Budget", "Bond Actor Salary"]
for col in columns:
    bond_df[col] = bond_df[col].apply(convert_to_string_and_add_millions)
```

```python
bond_df
```
