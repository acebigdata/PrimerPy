```python
## Function to convert wide to long
def gather( df, key, value, cols ):
    id_vars = [ col for col in df.columns if col not in cols ]
    id_values = cols
    var_name = key
    value_name = value
    return pd.melt( df, id_vars, id_values, var_name, value_name )
```

```python
ava_rev=pd.read_csv("ava_reve.csv", encoding='latin-1')
```

```python
wide_col_names = list(ava_rev.columns)
wide_col_names
```

```python
wide_col_names.index("FY1998")
```

```python
wide_to_long_cols = wide_col_names[wide_col_names.index("FY1998"):]
wide_to_long_cols
```

```python
ava_rev_long = gather(ava_rev, 'Year', 'Total Revenue', wide_to_long_cols)
```

```python
ava_rev_long
```

```python
ava_rev_long.info()
```

```python
ava_rev_long["Region"] = ava_rev_long["Region"].astype("category")
ava_rev_long["Country"] = ava_rev_long["Country"].astype("category")
ava_rev_long["Primary Industry"] = ava_rev_long["Primary Industry"].astype("category")
ava_rev_long["AV Industry"] = ava_rev_long["AV Industry"].astype("category")
ava_rev_long["AV Sector"] = ava_rev_long["AV Sector"].astype("category")
ava_rev_long["Year"] = ava_rev_long["Year"].astype("category")
```

```python
ava_rev_long["Total Revenue"] = ava_rev_long["Total Revenue"].apply(lambda x : x.replace("$", "").replace(",",""))
```

```python
ava_rev_long_clean = ava_rev_long[ava_rev_long["Total Revenue"] != "(Invalid Identifier) "]
```
