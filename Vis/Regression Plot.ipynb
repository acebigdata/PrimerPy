```python
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset("tips")
```

```python
tips.head()
```

```python
sns.lmplot(x = "total_bill", y = "tip", data = tips, hue = "sex", markers = ["v", "o"], 
           scatter_kws={"s": 50})
```

```python
sns.lmplot(x = "total_bill", y = "tip", data = tips, col = "time", row = "day", hue = "sex")
```
