```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
data_iris = sns.load_dataset("iris")
data_iris.head()
```

```python
sns.pairplot(data_iris)
```

```python
data_pair = sns.PairGrid(data_iris)
data_pair.map(plt.scatter)
```

```python
data_pair = sns.PairGrid(data_iris)
data_pair.map_diag(sns.distplot)
data_pair.map_upper(plt.scatter)
data_pair.map_lower(sns.kdeplot)
```

### Facet grid

```python
tips = sns.load_dataset("tips")
tips.head()
```

```python
tips_grid = sns.FacetGrid(tips, col = "time", row = "smoker")
tips_grid.map(plt.scatter, x = "total_bill", y = "tip", data = tips)
```
