```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
tips = sns.load_dataset("tips")
tips.head()
```

```python
sns.set_style("ticks")
sns.countplot(x = "sex", data = tips)
sns.despine()
```

```python
sns.lmplot(x = "total_bill", y="tip", data = tips, hue = "sex", palette="seismic")
```
